```python
df_all.loc[df_all['콜 성공 여부'] == True, 'success_yn'] = 'y'
df_all.loc[df_all['콜 성공 여부'] != True, 'success_yn'] = 'n'
```

## 전사 해피콜 작업시트

```python
전사_판토모나 = pd.read_clipboard()
```

```python
전사_판토모나
```

```python
전사_써큐시안 = pd.read_clipboard()
```

```python
전사_써큐시안
```

```python
전사_list = [전사_판토모나, 전사_써큐시안]
```

```python
전사_통합 = pd.concat(전사_list, ignore_index=True)
```

```python
전사_통합
```

```python
전사_통합.loc[전사_통합['1차 시도'] == 'TRY', 'try_yn'] = 'y'
전사_통합.loc[전사_통합['1차 시도'] != 'TRY', 'try_yn'] = 'n'

전사_통합.loc[전사_통합['콜 성공 여부'] == 'Y', 'success_yn'] = 'y'
전사_통합.loc[전사_통합['콜 성공 여부'] != 'Y', 'success_yn'] = 'n'`
```

```python
전사_통합.loc[전사_통합['구매 제품'].str.contains('써큐시안'), 'nick'] = '써큐시안블러드케어'
전사_통합.loc[전사_통합['구매 제품'].str.contains('판토모나맨'), 'nick'] = '판토모나남성용'
전사_통합.loc[전사_통합['구매 제품'].str.contains('판토모나레이디'), 'nick'] = '판토모나여성용'
전사_통합.loc[전사_통합['구매 제품'].str.contains('판토모나하이퍼포머'), 'nick'] = '판토모나하이퍼포머'
전사_통합.loc[전사_통합['구매 제품'] == '판토모나', 'nick'] = '판토모나하이퍼포머'
```

```python
전사_통합.columns
```

```python
전사_통합['통화시간'].fillna('0:00:00', inplace=True)

# 시간 문자열을 시, 분, 초로 분해
전사_통합['hours'] = 전사_통합['통화시간'].str.split(':').str[0].astype(int)
전사_통합['minutes'] = 전사_통합['통화시간'].str.split(':').str[1].astype(int)
전사_통합['seconds'] = 전사_통합['통화시간'].str.split(':').str[2].astype(int)

# 시, 분, 초를 모두 초로 변환하고 합산
전사_통합['call_time'] = 전사_통합['hours'] * 3600 + 전사_통합['minutes'] * 60 + 전사_통합['seconds']
```

```python
전사_통합[['해피콜날짜', 'nick', '연락처', 'try_yn', 'success_yn', 'call_time']].to_csv('call_log_2.csv', encoding='utf-8-sig', index=False)
```
