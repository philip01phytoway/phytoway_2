```python
import os
import pandas as pd
```

```python
directory = "C:/Users/user/Desktop/데이터분석팀/데이터수집/연관검색어/"

files = os.listdir(directory)

df1 = pd.DataFrame()

for file_ in files:
    df2 = pd.read_excel(directory + file_)
    df2['메인키워드'] = file_[0:-5]
    df1 = pd.concat([df1, df2])

print(df1)
```

```python
df1['메인키워드'].unique()
```

```python
df_drop_duplicates = df1.drop_duplicates(subset='연관키워드', keep='first') 
```
